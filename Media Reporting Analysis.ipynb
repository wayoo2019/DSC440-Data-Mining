```python
import pandas as pd
import numpy as np

D_2020 = pd.read_excel('2020 Police Shooting.xlsx', index_col=0)  
D_2019 = pd.read_excel('2019 Police Shooting.xlsx', index_col=0) 
D_2018 = pd.read_excel('2018 Police Shooting.xlsx', index_col=0) 
D_2017 = pd.read_excel('2017 Police Shooting.xlsx', index_col=0) 
D_2016 = pd.read_excel('2016 Police Shooting.xlsx', index_col=0) 
```

```python
frames = [D_2020, D_2019, D_2018, D_2017, D_2016]
D = pd.concat(frames)
D = D[D['Source'].notna()]
D_new= D[D['Race'].notna()]
D_new.info()
```

```python
D_new['Race'].value_counts()
```

```python
media = D_new['Source']
media = [x.replace(",", "") for x in media]
for i in range(len(media)):
    media[i] = list(media[i].split("\xa0"))
media
```

```python
media_list = []
for m in media:
    for outlet in m:
        media_list.append(outlet)
media_set = set(media_list)
D_media = pd.DataFrame(media_list)
D_media[0].value_counts()
```

```python
media_dic_race = {}
media_dic_f = {}
for media_name in media_set:
    reported_race = []
    for i in range(len(media)):
        if media_name in media[i]:
            reported_race.append(D_new['Race'][i])
    media_dic_race[media_name] = reported_race
    media_dic_f[media_name] = len(reported_race)
media_dic_f
```

```python
df = pd.DataFrame(list(media_dic_f.items()),columns = ['Media','Frequency'])
df = df.sort_values(by='Frequency', ascending=False)
```

```python
df.to_csv('media.csv')
```
